```python
import os
import time #gives us time to move around
import uuid #gives each picture a unique identifier
import cv2 #allows us to work with sensors and cameras
```

```python
IMAGES_PATH = os.path.join('data','images') #images going to go in folder called data
number_images = 30 #collect 30 images for samples
```

```python
cap = cv2.VideoCapture(0) #connecting to video capture device
#running loop to capture images
for imgnum in range(number_images):
    print('Collecting Image{}'.format(imgnum))
    #read from our capture device
    ret, frame = cap.read()
    #naming images
    imgname = os.path.join(IMAGES_PATH,f'{str(uuid.uuid1())}.jpg')
    cv2.imwrite(imgname,frame)
    cv2.imshow('frame',frame)
    #0.5 sec in between every picture
    time.sleep(0.5)

    #loop end case
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
